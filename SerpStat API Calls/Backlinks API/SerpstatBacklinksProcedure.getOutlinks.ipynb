```python
from datetime import datetime
import pandas as pd
import requests
from pandas import json_normalize
```

* The SerpstatBacklinksProcedure.getOutlinks method gives you a list of external backlinks of a site you analyze. 
* The data set is similar to the Backlinks report.
* Use case: with the SerpstatBacklinksProcedure.getOutlinks method, check the sites connected to the competitor's domain so you could establish a connection with these sites too to promote your site.

```python
api_token = '185b0a40fdd3abccb33f353c8e3e2918'                                                         
api_url_pattern = 'https://api.serpstat.com/v{version}?token={token}'
api_url = api_url_pattern.format(version=4, token=api_token)

headers = {
    "Content-Type": "application/json",
    "X-Api-Key": f"{api_token}",
}

data ={
 "id": 1,
 "method": "SerpstatBacklinksProcedure.getOutlinks",
 "params": {
     "query": "prosites.com",
     "searchType": "domain",
     "sort": "url_from",
     "order": "desc",
     "page": 1,
     "size": 2,
     "linkPerDomain": 1,
     "complexFilter": [
            [
                {
                    "field": "url_from",
                    "compareType": "notContains",
                    "value": [
                        "7654765"
                    ]
                },
                {
                    "field": "link_nofollow",
                    "compareType": "contains",
                    "value": [
                        "follow",
                        "nofollow",
                        "ugc",
                        "sponsored"
                    ]
                },
                {
                    "field": "anchor",
                    "compareType": "notContains",
                    "value": [
                        "лопспрол"
                    ]
                },
                {
                    "field": "check",
                    "compareType": "lte",
                    "value": [
                        "1.1.2030"
                    ]
                },
                {
                    "field": "links_external",
                    "compareType": "between",
                    "value": [
                        "1",
                        "500"
                    ]
                }
            ],
            [
                {
                    "field": "links_external",
                    "compareType": "lte",
                    "value": [
                        "800"
                    ]
                }
            ]
        ]
  }
}

api_resp = requests.post(api_url, json=data, headers=headers)
json_response = api_resp.json()
sliced_json = json_response['result']['data']
df = pd.json_normalize(sliced_json)
df
```
