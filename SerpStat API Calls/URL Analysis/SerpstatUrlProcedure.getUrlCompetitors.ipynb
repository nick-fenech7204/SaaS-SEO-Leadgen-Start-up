* The SerpstatUrlProcedure.getUrlCompetitors method returns the list of URL competitors to the queried URL. 
* The data set is similar to the Competitors report in the URL analysis.

```python
api_token = '185b0a40fdd3abccb33f353c8e3e2918'                                                         
api_url_pattern = 'https://api.serpstat.com/v{version}?token={token}'
api_url = api_url_pattern.format(version=4, token=api_token)

headers = {
    "Content-Type": "application/json",
    "X-Api-Key": f"{api_token}",
}

data = {
    "id": "1",
    "method": "SerpstatUrlProcedure.getUrlCompetitors",
    "params": {
     "se": "g_us",
     "url": "https://www.prosites.com/dental-marketing/",
     "sort": {"cnt": "desc"},
     "page": "1",
     "size": "10"
 }
}

api_resp = requests.post(api_url, json=data, headers=headers)
json_response = api_resp.json()
sliced_json = json_response['result']['data']
df = pd.json_normalize(sliced_json)
df
## Contains Domain Rank 
```
