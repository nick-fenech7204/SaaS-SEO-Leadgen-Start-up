```python
from datetime import datetime
import pandas as pd
import requests
from pandas import json_normalize
```

* The SerpstatKeywordProcedure.getCompetitors method lists the domains that rank for the given keyword in Google top-20 results. 
* The data set is similar to the Keyword research — SEO research — Competitors report.

```python
api_token = '185b0a40fdd3abccb33f353c8e3e2918'                                                         
api_url_pattern = 'https://api.serpstat.com/v{version}?token={token}'
api_url = api_url_pattern.format(version=4, token=api_token)

headers = {
    "Content-Type": "application/json",
    "X-Api-Key": f"{api_token}",
}

data = {
    "id": "1",
    "method": "SerpstatKeywordProcedure.getCompetitors",
    "params": {
    "keyword": "9MM pistol",
    "se": "g_us",
    "sort": {"traff":"desc"},
    "size": 20
 }
}

api_resp = requests.post(api_url, json=data, headers=headers)
json_response = api_resp.json()
sliced_json = json_response['result']['data']
df = pd.DataFrame.from_dict(sliced_json, orient='index')
df.reset_index(drop=False, inplace=True)
df.rename(columns={'index': 'domain'}, inplace=True)
df
## Contains Domain Rank 
```
