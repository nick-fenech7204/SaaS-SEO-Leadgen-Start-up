* The SerpstatKeywordProcedure.getKeywordFullTop method shows you Google's top-100 search results for the analyzed keyword. 
* The data set is similar to the Top by keyword report in Keyword research.

```python
api_token = '185b0a40fdd3abccb33f353c8e3e2918'                                                         
api_url_pattern = 'https://api.serpstat.com/v{version}?token={token}'
api_url = api_url_pattern.format(version=4, token=api_token)

headers = {
    "Content-Type": "application/json",
    "X-Api-Key": f"{api_token}",
}

data = {
        "id": 1,
        "method": "SerpstatKeywordProcedure.getKeywordFullTop",
        "params": {
        "keyword": "dentist",
        "se": "g_us",
        "sort": {
        "position": "asc"
                },
        "size": 50
    }
}

api_resp = requests.post(api_url, json=data, headers=headers)
json_response = api_resp.json()
sliced_json = json_response['result']['top']
df = pd.json_normalize(sliced_json)
df
## Contains SDR
```
