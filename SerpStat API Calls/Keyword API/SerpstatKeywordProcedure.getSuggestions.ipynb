* The SerpstatKeywordProcedure.getSuggestions method 
* Shows search suggestions for the keyword you requested (they are found by the full-text search). 
* The data set is similar to the Keyword research — SEO research — Search suggestions report.

```python
api_token = '185b0a40fdd3abccb33f353c8e3e2918'                                                         
api_url_pattern = 'https://api.serpstat.com/v{version}?token={token}'
api_url = api_url_pattern.format(version=4, token=api_token)

headers = {
    "Content-Type": "application/json",
    "X-Api-Key": f"{api_token}",
}

data = {
 "id": "1",
 "method": "SerpstatKeywordProcedure.getSuggestions",
 "params": {
   "keyword": "dental",
   "se": "g_us",
   "page": "1",
   "size": "20"
  }
 }

api_resp = requests.post(api_url, json=data, headers=headers)
json_response = api_resp.json()
sliced_json = json_response['result']['data']
df = pd.json_normalize(sliced_json)
df
```
