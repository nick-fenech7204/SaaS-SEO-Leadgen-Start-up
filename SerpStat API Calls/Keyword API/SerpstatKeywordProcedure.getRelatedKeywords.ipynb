```python
from datetime import datetime
import pandas as pd
import requests
from pandas import json_normalize
```

* The SerpstatKeywordProcedure.getRelatedKeywords method 
* Shows all search queries that are semantically related to the searched keyword. 
* The data set is similar to the Keyword research / Related keywords report.

```python
api_token = '185b0a40fdd3abccb33f353c8e3e2918'                                                         
api_url_pattern = 'https://api.serpstat.com/v{version}?token={token}'
api_url = api_url_pattern.format(version=4, token=api_token)

headers = {
    "Content-Type": "application/json",
    "X-Api-Key": f"{api_token}",
}

data = {
    "id": 1,
    "method": "SerpstatKeywordProcedure.getRelatedKeywords",
    "params": {
        "keyword": "guns near me",
        "se": "g_us",
        "sort": {"region_queries_count":"desc"},
        "page": 1,
        "size": 10
    }
}

api_resp = requests.post(api_url, json=data, headers=headers)
json_response = api_resp.json()
sliced_json = json_response['result']['data']
df = pd.json_normalize(sliced_json)
df
```
